```python
import yt_dlp
```

```python
# SEARCH
def search_song(query):

    with yt_dlp.YoutubeDL({"quiet": True}) as ydl:
        info = ydl.extract_info(f"ytsearch5:{query}", download=False)
        print(info["entries"][0]["title"])

    return info["entries"]
```

```python
search_song(input("search: "))
```
